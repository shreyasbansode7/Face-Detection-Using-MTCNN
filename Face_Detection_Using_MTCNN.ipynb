```python
from mtcnn.mtcnn import MTCNN
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pixels1 = plt.imread('/content/image1.jpg')
pixels2 = plt.imread('/content/image2.jpg')
pixels3 = plt.imread('/content/image3.jpg')
pixels4 = plt.imread('/content/image4.jpg')
pixels5 = plt.imread('/content/image5.jpg')
```

```python
detector = MTCNN()
```

```python
results1 = detector.detect_faces(pixels1)
results2 = detector.detect_faces(pixels2)
results3 = detector.detect_faces(pixels3)
results4 = detector.detect_faces(pixels4)
results5 = detector.detect_faces(pixels5)
```

```python
results1
results2
results3
results4
results5
```

```python
# draw an image with detected objects
def draw_facebox(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot the image
    plt.imshow(data)
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = plt.Rectangle((x, y), width, height, fill=False, color='green')
        # draw the box
        ax.add_patch(rect)

        # show the plot
    plt.show()
```

```python
draw_facebox('/content/image1.jpg', results1)
draw_facebox('/content/image2.jpg', results2)
draw_facebox('/content/image3.jpg', results3)
draw_facebox('/content/image4.jpg', results4)
draw_facebox('/content/image5.jpg', results5)
```

```python
# draw an image with detected objects
def draw_facebox_and_keypoints(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot the image
    plt.imshow(data)
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = plt.Rectangle((x, y), width, height,fill=False, color='orange')
        # draw the box
        ax.add_patch(rect)
        # draw the dots
        for key, value in result['keypoints'].items():
            # create and draw dot
            dot = plt.Circle(value, radius=1, color='red')
            ax.add_patch(dot)
            # show the plot
    plt.show()# filename = 'test1.jpg' # filename is defined above, otherwise uncomment
```

```python
draw_facebox_and_keypoints('/content/image1.jpg', results1)
draw_facebox_and_keypoints('/content/image2.jpg', results2)
draw_facebox_and_keypoints('/content/image3.jpg', results3)
draw_facebox_and_keypoints('/content/image4.jpg', results4)
draw_facebox_and_keypoints('/content/image5.jpg', results5)
```
